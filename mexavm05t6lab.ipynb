```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
gender = pd.read_csv('mexavm05t6lab - Gender.csv')
total = pd.read_csv('mexavm05t6lab - Total.csv')
agerange = pd.read_csv('mexavm05t6lab - AgeRange.csv')
education = pd.read_csv('mexavm05t6lab - Education.csv')
children = pd.read_csv('mexavm05t6lab - Children.csv')
income = pd.read_csv('mexavm05t6lab - Income.csv')
```

```python
gender.set_index('Gender', inplace=True)
total.set_index('Total', inplace=True)
agerange.set_index('AgeRange', inplace=True)
education.set_index('Education', inplace=True)
children.set_index('Children', inplace=True)
income.set_index('Income', inplace=True)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(25,1), dpi=300)
sns.set(font_scale=1.4)

sns.heatmap(total, cmap='YlOrRd', linewidth=0.3, annot=True, ax=ax, fmt='g', xticklabels=False)

ax.set_yticklabels(total.index, rotation=0)

#ax.set_ylabel('') 
ax.tick_params(axis='y', which='major', pad=100)

#ax3.set_xticklabels(gender.columns, rotation=30)
plt.show()
```
